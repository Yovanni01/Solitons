Solucion oscilador armónico simple

```python
w1=1
x_0=np.array([5,0]);
t_0=-2*np.pi;
N1=120;
h1=0.1;

# declaración de la función del sistema de EDOs
def f1(t1,r):
    s=np.zeros(2)
    s[0]=r[1]+0*t1; #solucion
    s[1]=-w1*w1*r[0]+0*t1; #Derivada
    return s

#Solucion numerica
x,t=cod.RK4V(N1,h1,f1,x_0,t_0);
#Solucion analitica
x_a=5*np.cos(w1*t);
#Velocidad analitica
v_a=-5*w1*np.sin(w1*t);

fig, ax = plt.subplots();
ax.plot(t,x[:,0],'co');
ax.plot(t,x[:,1],'yo');
ax.plot(t,x_a,'r-');
ax.plot(t,v_a,'b-');
ax.set_title('Oscilador armónico simple')
ax.set_ylim(-6,10);
ax.set_xlabel(r"$t$", fontsize=15, fontname='Times New Roman')
ax.set_ylabel(r"$x(t),v(t)$", fontsize=15, fontname='Times New Roman')
L=[r'$Solución-numérica$', r'$Velocidad-numérica$',r'$Solución-analítica$', r'$Velocidad-analítica$']
ax.legend(L)
plt.grid();
plt.show()
```

Ejercicio 3, soluciones estaticas con potencial de sombrero mexicano, kink

```python
l,m=1,1;
nu=m/np.sqrt(l);
N1=200;
h1=0.1;

#Kink
def f1(x1,phi1):
    s=-np.sqrt((l/2))*(phi1**2-nu**2); #Derivada   s=d(phi)/dx                                  
    return s

def Potencial(phi1):
    V=(l/4)*((phi1**2-nu**2)**2);
    return V

#Solucion numerica
y, x = cod.RK4(N1,h1,f1,-0.9999985573,-10);
#Solucion analitica
y_a=np.tanh(1/(np.sqrt(2))*x);
#Derivada
Dx_y=cod.Der(y,h1);
#Densidad de energia
pho=cod.Dens_E1(y,Dx_y,Potencial);
#Energia
E=cod.int(pho,h1);
#Energia analitica:
E_a=(2*np.sqrt(2)*m**3)/3*l;

print("El valor de la energía numérica es: E={:.4f}".format(E));
print("\n mientras que el valor de la energía analítica es: E={:.4f}".format(E_a));

fig, ax = plt.subplots();
ax.plot(x,y_a,'r-');
ax.plot(x,y,'c--');
ax.plot(x,pho,'g--');
ax.set_title('Solución estática, kink, sombrero mexicano')
ax.set_xlabel(r"$x$", fontsize=15, fontname='Times New Roman');
ax.set_ylabel(r"$\phi(x),\rho(x)$", fontsize=15, fontname='Times New Roman');
L=[r'$Solución-analítica$',r'Solución-numérica', r'$Densidad-de-energía$'];
plt.grid();
ax.legend(L);
plt.show();
```
